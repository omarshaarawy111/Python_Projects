## Description
> ##### The Hangman program randomly selects a secret word from a list of secret words.
> ##### The Game: Here, a random word (a fruit name) is picked up from our collection and the player gets limited chances to win the game.
> ##### When a letter in that word is guessed correctly, that letter position in the word is made visible. In this way,all letters of the word are to be guessed before all the chances are over.
> ##### For convenience, we have given length of word + 2 chances. For example, word to be guessed is mango, then user gets 5 + 2 = 7 chances, as mango is a five letter word.

```python
# project 2
# Libraries we need 
import random
# Words to play on 
def guess_level_1 (someWords):
    chances=0
    chances=int(chances)
    print()
    word = random.choice(someWords)
    print(word)
    print()
    while 1 :
            print()
            if chances < 3 :
                guessed_word=input("enter the guessed word ")
                if guessed_word == word :
                    print()
                    print("you guessed the right word and it is {}".format(word))
                    print()
                    break 
                else :
                    chances += 1 
                    print()
                    print("you guessed the wrong word , try again")
                    print()    
            elif chances == 3 :
                print()
                print("you finish all your chances  word and it is {} , good bye".format(word))
                print()
                break 
                
def guess_level_2 (someWords):
    chances=0
    chances=int(chances)
    print()
    word = random.choice(someWords)
    print(word)
    #buffer
    word_hang=""    
    print()
    while 1 :
            
                if chances < len(word)+2 :
                    print()
                    if set(word_hang) == set(word) :
                            print("you win and the word is {}".format(word))
                            print()
                            break 
                    guessed_char=input("enter the guessed letter :  ")
                    chances += 1 
                    if  len(guessed_char)==1 :                    
                        
                        for ch in word :
                            if  ch == guessed_char  or ch in word_hang :
                                word_hang=word_hang+ch
                                print(ch,end="")
                            else :
                                print("_",end="")
                            
                        
                    else : 
                        print()
                        print("enter letter only ")
                        print()
                        continue   
                else :    
                    print()
                    print("you finish all your chances  word and it is {} , good bye".format(word))
                    print()
                    break 
                               
             
    
def take () :
    someWords = '''apple banana mango strawberry
    orange grape pineapple apricot lemon coconut watermelon 
    cherry papaya berry peach lychee muskmelon'''
    someWords = someWords.split(' ')
    # randomly choose a secret word from our "someWords" LIST.
    try:
        choose=int(input("enter the level you want : "))
    except Exception as e :
        print(e)
        print()
        take()
        print()  
    if choose == 1 :
        print()
        print("Welcome to first level") 
        guess_level_1(someWords)
    elif choose == 2 :
        print()
        print("Welcome to second level") 
        guess_level_2(someWords)
    else :
        print("invalid choice")
        print()
        take()
        print()       
    char=input("do you want to play again ? [y/n] : ") 
    try :
        if char == 'y' :
            print()
            print("Thanks for playing")
            print()
            take()
        elif char == 'n' :
            print()
            print("Thanks for your time")
            print()
    except Exception as e : 
        print()
        print(e)  
take()    

```
